```python
import pandas as pd
import numpy as np
data=pd.read_csv("Tips.csv")
da={
    'col1':data['Total Bill'],
    'col2':data['Tip'],
    'col3':data['Bill Per Person'],
}
df=pd.DataFrame(da)
z_scores=np.abs(df-df.mean())/df.std()
threshold=3
outlier=z_scores>threshold
df_no_duplicate=df[(z_scores<threshold).all(axis=1)]
df_no_duplicate
```

```python
lower_quantile=0.05
upper_quantile=0.95
lower_bound=df.quantile(lower_quantile)
upper_bound=df.quantile(upper_quantile)
df_capped=df.clip(lower_bound,upper_bound,axis=1)
print("\nQuantile Values0.05 and 0.95")
print("lower bound(0.05)is:")
print(lower_bound)
print("\nUpper bound(0.95):")
print(upper_bound)
df_capped=df.clip(lower_bound,upper_bound,axis=1)
print("original")
print(df)
print("after capping")
print(df_capped)
```

```python
df_log_transform=np.log1p(df)
print("original")
print(df)
print("\nafter log")
print(df_log_transform)
```

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
def replace_outliers_with_median(df):
    df_imputed=df.copy()
    for column in df.columns:
        median=df[column].median()
        df_imputed[column]=df_imputed[column].apply(lambda x: median if x < lower_bound[column] or x > upper_bound[column] else x)
        return df_imputed
df_imputed=replace_outliers_with_median(df)
print("original data")
print(df)
print("replacing outliers")
print(df_imputed)
```
